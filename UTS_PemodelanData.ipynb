```python
import pandas as pd 
import numpy as np
import plotly.express as px
```

```python
file = '/content/drive/MyDrive/DataSets/datauts.csv'
data = pd.read_csv(file)
```

```python
data
```

```python
data = data.drop(columns=['Lat','Long'])
data.head()
```

```python
datanew = data.drop(columns=['Province/State'])
datanew = data.groupby(by='Country/Region').aggregate(np.sum).T
datanew
```

```python
datanew.index.name ='Date'
datanew = datanew.reset_index()
datanew.tail()
```

```python
datanew['Date'] = pd.to_datetime(datanew['Date'])
datanew
```

```python
datanew['Date'] = datanew['Date'].dt.strftime('%m/%d/%Y')
```

```python
data_indo = datanew[['Date','Indonesia']]
# data_china = data_china.drop(columns=['Province/State'])
data_indo
```

```python
datanew = datanew.melt(id_vars='Date').copy()
datanew
```

```python
datanew.rename(columns={'value':'Confirmed'}, inplace=True)
datanew.head()
```

Max & Min Tanggal

```python
max_date= datanew['Date'].max()
max_date
```

```python
min_date= datanew['Date'].min()
min_date
```

**Kasus Tertinggi Global**

```python
max = datanew['Confirmed'].max()
max
```

Jumlah Total Kasus Global

```python
total_confirmed = datanew['Confirmed'].sum()
total_confirmed
```

**Grafik Kasus Pada Setiap Negara**

```python
fig = px.bar(datanew,x='Country/Region', y='Confirmed')
fig.show()
```

```python
fig = px.bar(datanew.sort_values('Confirmed',ascending=False).head(10), x='Country/Region', y='Confirmed',text='Confirmed')
fig.show()
```

```python
fig = px.scatter(datanew, x='Date',y='Confirmed', color='Country/Region')
fig.show()
```
